```python
#Importing packages
import pandas as pd
import os
import numpy as np
```

```python
#Loading raw data CSV: 2016 Carbon Monoxide (CO)
file_to_load = ("2016-co-sfbay.csv")
CO_2016 =pd.read_csv(file_to_load)
CO_2016.head()
```

```python
#Checking Columns
CO_2016.columns
```

```python
#Checking Null Values
CO_2016.columns.isnull()
```

```python
#Dropping Columns
CO_2016 = CO_2016.drop(columns=['Source','Site ID','POC','DAILY_OBS_COUNT','PERCENT_COMPLETE','AQS_PARAMETER_CODE','AQS_PARAMETER_DESC', 'CBSA_CODE','CBSA_NAME','STATE_CODE','STATE','COUNTY_CODE','COUNTY','SITE_LATITUDE', 'SITE_LONGITUDE'], axis=1)
```

```python
#Checking Columns after drop
CO_2016.columns
```

```python
#Checking DataFrame
CO_2016.head()
```

```python
#Renaming Columns
CO_2016 = CO_2016.rename(columns={"Site Name": "Site_Name"})
CO_2016 = CO_2016.rename(columns={"Daily Max 8-hour CO Concentration": "Daily_Max_8-hr_CO_Concentration"})
CO_2016.head()
```

```python
#Checking Count: 
CO_2016.count()
```

```python
#Filtering on Site_Name for: San Francisco (SF)
CO_2016 = CO_2016[(CO_2016.Site_Name == "San Francisco")]
CO_2016.head()
```

```python
#Checking SF Count: 359
CO_2016.count()
```

```python
CO_2016.to_csv(r'2016_co.csv', header=True, index=False)
```
