```python
#Importing packages
import pandas as pd
import os
import numpy as np
import glob
```

```python
#path = "/home/user/data/"
path = "/Users/selenajdeleon/Desktop/Analysis_Projects/Final_Project/ML---Final-Project---Pollution-Future-Impact-Analysis-/clean_data"

#Selecting all PM2.5 CSV files
all_files = glob.glob(os.path.join(path, "20**_pm2_5.csv"))

#Sorting files by 'Date' column (asc)
all_files.sort()

#Merging 2015-2020 PM2.5 files
df_from_each_file = [pd.read_csv(f, sep=',') for f in all_files]
df_from_each_file
df_merged   = pd.concat(df_from_each_file)

#Creating CSV file for merged PM2.5 files (2015-2020)
df_merged.to_csv("pm2_5.csv", header = True, index = False)
```
