```python
#Importing packages
import pandas as pd
import os
import numpy as np
import glob
```

```python
#path = "/home/user/data/"
path = "/Users/selenajdeleon/Desktop/Analysis_Projects/Final_Project/ML---Final-Project---Pollution-Future-Impact-Analysis-/clean_data"

#Selecting all OZ CSV files
all_files = glob.glob(os.path.join(path, "20**_oz.csv"))

#Sorting files by 'Date' column (asc)
all_files.sort()

#Merging 2015-2020 OZ files
df_from_each_file = [pd.read_csv(f, sep=',') for f in all_files]
df_from_each_file
df_merged   = pd.concat(df_from_each_file)

#Creating CSV file for merged OZ files (2015-2020)
df_merged.to_csv("oz.csv", header = True, index = False)
```

```python
#Renaming DAILY_AQI_VALUE column to include name of pollutant for ML model
df_merged = df_merged.rename(columns={"DAILY_AQI_VALUE": "AQI_O3"})
df_merged.head()
```
