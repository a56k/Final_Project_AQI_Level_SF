```python
#Importing packages
import pandas as pd
import os
import numpy as np
```

```python
#Loading raw data CSV: 2017 Carbon Monoxide (CO)
file_to_load = ("2017-co-sfbay.csv")
CO_2017 =pd.read_csv(file_to_load)
CO_2017.head()
```

```python
#Checking Columns
CO_2017.columns
```

```python
#Checking Null Values
CO_2017.columns.isnull()
```

```python
#Dropping Columns
CO_2017 = CO_2017.drop(columns=['Source','Site ID','POC','DAILY_OBS_COUNT','PERCENT_COMPLETE','AQS_PARAMETER_CODE','AQS_PARAMETER_DESC', 'CBSA_CODE','CBSA_NAME','STATE_CODE','STATE','COUNTY_CODE','COUNTY','SITE_LATITUDE', 'SITE_LONGITUDE'], axis=1)
```

```python
#Checking Columns after drop
CO_2017.columns
```

```python
#Checking DataFrame
CO_2017.head()
```

```python
#Renaming Columns
CO_2017 = CO_2017.rename(columns={"Site Name": "Site_Name"})
CO_2017 = CO_2017.rename(columns={"Daily Max 8-hour CO Concentration": "Daily_Max_8-hr_CO_Concentration"})
CO_2017.head()
```

```python
#Checking Count: 3506
CO_2017.count()
```

```python
#Filtering on Site_Name for: San Francisco (SF)
CO_2017 = CO_2017[(CO_2017.Site_Name == "San Francisco")]
CO_2017.head()
```

```python
#Checking SF Count: 349
CO_2017.count()
```

```python
CO_2017.to_csv(r'2017_co.csv', header=True, index=False)
```
