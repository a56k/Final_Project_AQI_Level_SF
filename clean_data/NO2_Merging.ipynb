```python
#Importing packages
import pandas as pd
import os
import numpy as np
import glob
```

```python
#Path = "/home/user/data/"
path = "/Users/selenajdeleon/Desktop/Analysis_Projects/Final_Project/ML---Final-Project---Pollution-Future-Impact-Analysis-/clean_data"

#Selecting all NO2 CSV files
all_files = glob.glob(os.path.join(path, "20**_no2.csv"))

#Sorting files by 'Date' column (asc)
all_files.sort()

#Merging 2015-2020 NO2 files
df_from_each_file = [pd.read_csv(f, sep=',') for f in all_files]
df_from_each_file
df_merged   = pd.concat(df_from_each_file)

#Creating CSV file for merged NO2 files (2015-2020)
df_merged.to_csv("no2.csv", header = True, index = False)
```
