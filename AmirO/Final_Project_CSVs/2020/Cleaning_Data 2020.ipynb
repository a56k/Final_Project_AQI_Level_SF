```python
#Import dependencies
import pandas as pd
import os
import numpy as np
```

```python
#Load CSV
file_to_load = ("SF 2020 CO.csv")
CO_2020 =pd.read_csv(file_to_load)
CO_2020.head()
```

```python
#Check Columns
CO_2020.columns
```

```python
#Check Null Values
CO_2020.columns.isnull()
```

```python
#Drop Columns
CO_2020= CO_2020.drop(columns=['Source','Site ID','POC','AQS_PARAMETER_CODE','DAILY_OBS_COUNT','PERCENT_COMPLETE','AQS_PARAMETER_DESC', 'CBSA_CODE','CBSA_NAME','STATE_CODE','STATE','COUNTY_CODE', 'SITE_LATITUDE', 'SITE_LONGITUDE','COUNTY'], axis=1)
```

```python
#Check DataFrame
CO_2020.head()
```

```python
#Rename Site Name & Daily Concentration Columns
CO_2020 = CO_2020.rename(columns={"Site Name": "Site_Name"})
CO_2020 = CO_2020.rename(columns={"Daily Max 8-hour CO Concentration":"Daily_Max_8-hr_CO_Concentration"})
CO_2020.head()
```

```python
#Filter for SF
#CO_2020 = CO_2020[(CO_2020.Site_Name == "Oakland West") | (CO_2020.Site_Name == "San Francisco")]
CO_2020 = CO_2020[(CO_2020.Site_Name == "San Francisco")]
CO_2020.head(10)
```

```python
#Check Count//LEAP YEAR
CO_2020.count()
```

```python
#To CSV
CO_2020.to_csv(r'2020_co.csv', header= True, index=False)
```
