```python
#Import dependencies
import pandas as pd
import os
import numpy as np
```

```python
#Load CSV
file_to_load = ("SF 2019 CO.csv")
CO_2019 =pd.read_csv(file_to_load)
CO_2019.head()
```

```python
#Check Columns
CO_2019.columns
```

```python
#Check Null Values
CO_2019.columns.isnull()
```

```python
#Drop Columns
CO_2019 = CO_2019.drop(columns=['Source','Site ID','POC','AQS_PARAMETER_CODE','DAILY_OBS_COUNT','PERCENT_COMPLETE','AQS_PARAMETER_DESC', 'CBSA_CODE','CBSA_NAME','STATE_CODE','STATE','COUNTY_CODE', 'SITE_LATITUDE', 'SITE_LONGITUDE'], axis=1)
```

```python
#Check DataFrame
CO_2019.head()
```

```python
#Rename Site Name Column
CO_2019 = CO_2019.rename(columns={"Site Name": "Site_Name"})
CO_2019.head()
```

```python
#Filter for SF & Oakland West Count 730 is total
CO_2019 = CO_2019[(CO_2019.Site_Name == "Oakland West") | (CO_2019.Site_Name == "San Francisco")]
CO_2019.head(10)
```

```python
#Check Count for SF & Oakland West
CO_2019.count()
```

```python
#To CSV
CO_2019.to_csv(r'CO_2019_clean.csv', header= True, index=False)
```
