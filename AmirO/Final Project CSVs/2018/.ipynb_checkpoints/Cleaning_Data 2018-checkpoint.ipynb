```python
#Import dependencies
import pandas as pd
import os
import numpy as np
```

```python
#Load CSV
file_to_load = ("SF 2018 CO Original.csv")
CO_2018 =pd.read_csv(file_to_load)
CO_2018.head()
```

```python
#Check Columns
CO_2018.columns
```

```python
#Check Null Values
CO_2018.columns.isnull()
```

```python
#Drop Columns
CO_2018 = CO_2018.drop(columns=['Source','Site ID','POC','AQS_PARAMETER_CODE','AQS_PARAMETER_DESC', 'CBSA_CODE','STATE_CODE','STATE','COUNTY_CODE'], axis=1)
```

```python
#Check DataFrame
CO_2018.head()
```

```python
#Rename Column
CO_2018 = CO_2018.rename(columns={"Site Name": "Site_Name"})
CO_2018.head()
```

```python
#Check Count
CO_2018.count()
```

```python
#Filter for SF & Oakland
CO_2018 = CO_2018[(CO_2018.Site_Name == "Oakland") | (CO_2018.Site_Name == "San Francisco")]
CO_2018.head(10)
```

```python
#Check SF & Oakland Count 730 is total
CO_2018.count()
```
